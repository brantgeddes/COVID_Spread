```python
# Imports
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load data
trainData = pd.read_csv('./dataset/train.csv', index_col=0)
testData = pd.read_csv('./dataset/test.csv', index_col=0)
```

```python
# Pull data out

# Saskatchewan Stats
plotData = trainData[(trainData['Country_Region'] == "Canada") & (trainData['Province_State'] == "Saskatchewan") & (trainData['Target'] == "ConfirmedCases")]

# US Stats (Needs some additional work)
# plotData = trainData[(trainData['Country_Region'] == "US") & (trainData['Target'] == "ConfirmedCases")]

# World Stats (Chart scale is bad for daily vs summed)
# plotData = trainData
print("Number of Datapoints: ", plotData.shape[0], "\n")
```

```python
plotData["TargetSummed"] = plotData["TargetValue"].cumsum()
```

```python
# Plot data

sns.lineplot(x="Date", y="TargetValue", data=plotData)
plot = sns.lineplot(x="Date", y="TargetSummed", data=plotData)
plt.xlabel("Date")
plt.ylabel("Value")
plot.set_xticks(range(1))
plot.set_xticklabels(labels=[''], rotation=45, ha='right')
```
